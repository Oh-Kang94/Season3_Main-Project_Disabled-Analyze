```r
#   KNN을 위한 Training Data 준비
data.training <- subset(data.tr, select = -target)
data.training.label <- data.tr$target
```

```r
#   문제 및 정답 만들기
x   <- subset(data.test, select = -target)
y   <- data.test$target
```

# 학습시키기

```r
#  랜덤포레스트 학습 시키기
rf <- randomForest(target ~., data.tr, ntree=500)
```
